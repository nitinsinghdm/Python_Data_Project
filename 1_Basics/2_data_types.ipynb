#Built-in Types
Python has the following data types built-in by default:

| Type      | Data Type   | Example |
|-----------|------------|---------|
| Text      | str        | `"Data Nerd!"` |
| Numeric   | int        | `42` |
|           | float      | `3.14159` |
|           | complex    | `1 + 2j` |
| Sequence  | list       | `[1, 2, 3]` |
|           | tuple      | `(1, 2, 3)` |
|           | range      | `range(10)` |
| Mapping   | dict       | `{"key": "value"}` |
| Set       | set        | `{1, 2, 3}` |
|           | frozenset  | `frozenset([1, 2, 3])` |
| Boolean   | bool       | `True` or `False` |
| Binary    | bytes      | `b"Data"` |
|           | bytearray  | `bytearray(5)` |
|           | memoryview | `memoryview(b"Data")` |
| None      | NoneType   | `None` |
