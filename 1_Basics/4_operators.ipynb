#**Arithmetic Operators**

Arithmetic operators are used with objects (mainly numeric ones) to perform math operations:

| Operator | Name            | Example  |
|----------|----------------|----------|
| +        | Addition        | `x + y`  |
| -        | Subtraction     | `x - y`  |
| *        | Multiplication  | `x * y`  |
| /        | Division        | `x / y`  |
| %        | Modulus         | `x % y`  |
| **       | Exponentiation  | `x ** y` |
| //       | Floor Division  | `x // y` |


```python
2 + 3.0
```

```python
"What's " + 'Up ' + 'People!'
```

```python
"What's Up People! " * 5
```

```python
print('5 divided by 2 is', 5 / 2, 'with a remainder of', 5 % 2)
```

**Order of *Operations***
*  **Parentheses:** Anything in parentheses is executed first
*  **Exponents:** Operations involving exponents (including roots, indicated as fractional exponents)
*  **Multiplication and Division:** Processed from left to right
*  **Addition and Subtraction:** Also processed from left to right




```python
min_salary = 20_000 # _ is ignored by the interpreter for big numbers
max_salary = 100_000

avg_salary = (min_salary + max_salary) / 2
print(avg_salary)
```

# **Comparison Operators**

Comparison operators compare two values and determine the relation between them.

| Operator | Name                        | Example   |
|----------|----------------------------|-----------|
| ==       | Equal to                    | `x == y`  |
| !=       | Not equal to                | `x != y`  |
| >        | Greater than                | `x > y`   |
| <        | Less than                   | `x < y`   |
| >=       | Greater than or equal to    | `x >= y`  |
| <=       | Less than or equal to       | `x <= y`  |


```python
salary_nitin = 100_000
salary_sujeeth = 70_000
```

```python
salary_nitin == salary_sujeeth
```

```python
salary_sujeeth >= salary_nitin
```

```python
salary_nitin != salary_sujeeth
```

```python
if(salary_nitin>salary_sujeeth):
  print("Nitin's salary is greater than Sujeeth's")
```

## **Logical Operators**

Logical operators connect two or more conditions and evaluate the combined truth value.

| Operator | Name         | Example     |
|----------|-------------|------------|
| and      | Logical AND | `x and y`  |
| or       | Logical OR  | `x or y`   |
| not      | Logical NOT | `not x`    |


```python
job_wfh = True
job_remote = False
#job_no_degree = True

job_wfh and job_remote
```

```python
job_wfh = True
job_remote = False
job_no_degree = True

job_wfh or job_remote or not job_no_degree
```

```python
if job_wfh and job_remote or job_no_degree:
  print("You are qualified")
else:
  print("You are not qualified")
```

## **Identity Operators**

Identity operators check whether two variables refer to the same object in memory.

| Operator | Name         | Example         |
|----------|-------------|-----------------|
| is       | Identity     | `x is y`        |
| is not   | Not identity | `x is not y`    |


```python
job_skills = ['Python', 'Excel', 'SQL']
nitin_skills = job_skills

job_skills is nitin_skills
```

```python
job_skills = ['Python', 'Excel', 'SQL']
nitin_skills = ['Python', 'Excel', 'SQL']

job_skills is nitin_skills
```

## **Membership Operators**

Membership operators test whether a value exists inside a sequence
(such as strings, lists, tuples, sets, or dictionaries).

| Operator | Name            | Example        |
|----------|----------------|---------------|
| in       | Membership      | `x in y`      |
| not in   | Not membership  | `x not in y`  |


```python
'Job' in 'Job Skills'
```

```python
job_skills = ['Python', 'Excel', 'SQL']

'Python' in job_skills
```

```python
job_skills = ['Python', 'Excel', 'SQL']

'Python' not in job_skills
```

## **Bitwise Operators**

Bitwise operators perform operations on binary representations of integers.

| Operator | Name         | Example   |
|----------|-------------|----------|
| &        | Bitwise AND | `x & y`  |
| \|       | Bitwise OR  | `x | y`  |
| ^        | Bitwise XOR | `x ^ y`  |
| ~        | Bitwise NOT | `~x`     |
| <<       | Left Shift  | `x << y` |
| >>       | Right Shift | `x >> y` |


```python
bin(50)
```
