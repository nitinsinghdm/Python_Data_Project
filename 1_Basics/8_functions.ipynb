# **Lambda Function**

```python
salary_list = [600000, 2980000, 340000, 440000, 900000, 990000]

#def calculate_salary(base_salary, bonus_rate = .5):
#  return total_salary

total_salary_list = [(lambda x: x*.1)(salary) for salary in salary_list]

total_salary_list
```

```python
# LOL btw you can also do this

total_salary_list = [salary for salary in salary_list]

total_salary_list
```

# **Data Filtering**

```python
# Find out jobs that have Python and are remote
jobs_data = [
    {'job_title': 'Data Scientist',  'job_skills': ['Python', 'Machine Learning'], 'remote': True},
    {'job_title': 'Data Analyst',  'job_skills': ['Excel', 'SQL'], 'remote': False},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'Keras'], 'remote': True},
    {'job_title': 'Software Developer', 'job_skills': ['Java', 'C++'], 'remote': True},
    {'job_title': 'Data Scientist', 'job_skills': ['R', 'Statistics'], 'remote': False}
]
```

```python
list(filter(lambda jobs: jobs['remote'] and 'Python' in jobs['job_skills'], jobs_data))
```
